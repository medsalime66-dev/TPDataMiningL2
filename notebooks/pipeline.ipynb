- Web Scraping

```python
import requests
import pandas as pd

url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url, verify=False)

data = response.json()

df_api = pd.DataFrame(data)

df_api.head()
```

- utulisation de API

```python
import requests
import pandas as pd

url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url)

data = response.json()

df_api = pd.DataFrame(data)

df_api.head()
```

```python
final_df = pd.concat([excel, movies, df_api], axis=0)

final_df.head()
```
